```python
raw = nltk.corpus.gutenberg.raw('carroll-alice.txt')
```

```python
def get_data(raw):
    corpus = [sentence.replace('\n', '') for sentence in raw.split('\n\n') if sentence]
    return corpus
```

```python
corpus = get_data(raw)
N = 30
half_window_size = 2
lr = 0.01
n_epoch = 300
print_epoch = 10
batch_size = 10
num_workers=2
shuffle=False
```

```python
word2vec, loss_list = word2vec_train(corpus, N=N, half_window_size=half_window_size,
                                     lr=lr, n_epoch=n_epoch, batch_size=batch_size,
                                     print_epoch=print_epoch, num_workers=num_workers,
                                     shuffle=shuffle)
```
